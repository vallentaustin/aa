```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca dataset
df = pd.read_csv("studentVle.csv")
```

```python
# Menampilkan informasi dasar tentang dataset
df.info()
```

```python
# Menampilkan 5 baris pertama data
df.head(5)
```

```python
# Menampilkan 5 baris pertama data
df.tail(5)
```

```python
# Statistik deskriptif untuk kolom numerik
df.describe()
```

```python
# Menghitung jumlah nilai yang hilang
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Visualisasi distribusi kolom "sum_click"
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="sum_click", bins=20, kde=True)
plt.title("Distribusi sum_click")
plt.xlabel("sum_click")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Visualisasi hubungan antara "sum_click" dan "date" menggunakan scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="date", y="sum_click")
plt.title("Hubungan antara date dan sum_click")
plt.xlabel("date")
plt.ylabel("sum_click")
plt.show()
```

```python
# Visualisasi hubungan antara "id_site" dan "sum_click" menggunakan scatterplot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="id_site", y="sum_click")
plt.title("Hubungan antara id_site dan sum_click")
plt.xlabel("id_site")
plt.ylabel("sum_click")
plt.show()
```

```python
# Visualisasi hubungan antara "sum_click" dan "code_module" menggunakan boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="code_module", y="sum_click")
plt.title("Hubungan antara code_module dan sum_click")
plt.xlabel("code_module")
plt.ylabel("sum_click")
plt.show()
```

```python
# Visualisasi hubungan antara "code_presentation" dan "sum_click" menggunakan boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="code_presentation", y="sum_click")
plt.title("Distribusi sum_click per code_presentation")
plt.xlabel("code_presentation")
plt.ylabel("sum_click")
plt.xticks(rotation=45)
plt.show()
```

```python
# menampilkan hasil korelasi antar fitur yang ada dalam dataset
pivot_table = df.pivot_table(values="sum_click", index="code_module", columns="code_presentation")
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt=".1f")
plt.title("Distribusi sum_click per code_module dan code_presentation")
plt.xlabel("code_presentation")
plt.ylabel("code_module")
plt.show()
```
