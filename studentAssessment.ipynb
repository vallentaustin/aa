```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
df = pd.read_csv("studentAssessment.csv")
```

```python
#Print 5 data awal
df.head(5)
```

```python
#Print 5 data akhir
df.tail(5)
```

```python
# Menampilkan detail tentang dataframe
df.info()
```

```python
# Menampilkan statistik deskriptif
df.describe()
```

```python
# Cek missing value
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Menghapus baris dengan missing value
df.dropna(inplace=True)
```

```python
#Melakukan Plotting menggunakan diagram batang
plt.figure(figsize=(8, 6))
sns.histplot(df['score'], kde=True, bins=10)
plt.title('Distribusi Skor')
plt.xlabel('Skor')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# menampilkan hasil korelasi antar fitur yang ada dalam dataset
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi antara Kolom')
plt.show()
```

```python
#Melakukan plotting menggunakan box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_banked', y='score', data=df)
plt.title('Boxplot Skor berdasarkan is_banked')
plt.xlabel('is_banked')
plt.ylabel('Skor')
plt.show()
```

```python
# menampilkan data plotting menggunakan histogram
plt.figure(figsize=(16, 12))
sns.countplot(x='id_assessment', data=df)
plt.title('Jumlah Pengumpulan berdasarkan id_assessment')
plt.xlabel('id_assessment')
plt.ylabel('Jumlah Pengumpulan')
plt.xticks(rotation=90)
plt.show()
```

```python
#Melakukan plotting menggunakan scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='date_submitted', y='score', data=df)
plt.title('Hubungan antara date_submitted dan Skor')
plt.xlabel('date_submitted')
plt.ylabel('Skor')
plt.show()
```
